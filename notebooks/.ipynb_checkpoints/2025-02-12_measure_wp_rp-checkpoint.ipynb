### Generate a $z$ distribution for the random catalog
#### First plot the cdf for the data

```python
cdf, bins, patches = plt.hist(tab_gcatlo_nosel['redshift_quaia'], cumulative = True, density = True, bins = 100)
plt.xlabel('$z$')
plt.show()
```

#### Try randomly sampling within bins

```python
# follow this tutorial https://www.andreaamico.eu/data-analysis/2020/03/02/hist_sampling.html
values = np.random.rand(N_randlo)
value_bins = np.searchsorted(cdf, values)
tab_randlo['redshift_quaia'] = np.random.rand(len(value_bins))*np.diff(bins)[0]+bins[value_bins]
```

```python
plt.hist(tab_gcatlo_nosel['redshift_quaia'], histtype = 'step', density = True, label = 'mock', bins = 100)
plt.hist(tab_randlo['redshift_quaia'], histtype = 'step', density = True, label = 'random', bins = 100)
plt.xlabel('$z$')
plt.legend()
plt.show()
```

#### Try interpolating the cdf so it's more smooth

```python
def recenter(bins):
    return 0.5*(bins[1:]+bins[:-1])
```

```python
# Compute the empirical cumulative distribution function (ECDF)
z = np.sort(tab_gcatlo_nosel['redshift_quaia'])
ecdf = np.arange(N_gcatlo) / (N_gcatlo - 1)

# Create an interpolation function for inverse transform sampling
inv_cdf = interp.interp1d(cdf, recenter(bins)) # this will fail because it doesn't start at 0 and end at 1
inv_ecdf = interp.interp1d(ecdf, z) # have to interpolate, can't do binning method because each point --> cdf probability

# Generate new random samples from the estimated distribution
tab_randlo['redshift_quaia'] = inv_ecdf(np.random.uniform(size = N_randlo))
```

```python
plt.plot(z, ecdf, label = 'ecdf')
plt.plot(recenter(bins), cdf, linestyle = '--', label = 'cdf')
plt.xlabel('$z$')
plt.legend()
plt.show()
```

```python
plt.hist(tab_gcatlo_nosel['redshift_quaia'], histtype = 'step', density = True, label = 'mock', bins = 100)
plt.hist(tab_randlo['redshift_quaia'], histtype = 'step', density = True, label = 'ecdf', bins = 100)
plt.xlabel('$z$')
plt.legend()
plt.show()
```

### Plot for sanity check

```python
pimax = 40.0
```

```python
DD_counts = np.load('../results/DD_counts_rmin{}_rmax{}_pimax{}.npy'.format(str(rmin), str(rmax), str(pimax)))
DR_counts = np.load('../results/DR_counts_rmin{}_rmax{}_pimax{}.npy'.format(str(rmin), str(rmax), str(pimax)))
RR_counts = np.load('../results/RR_counts_rmin{}_rmax{}_pimax{}.npy'.format(str(rmin), str(rmax), str(pimax)))
```

```python
DD_counts['rpavg']  = np.mean([DD_counts['rmin'], DD_counts['rmax']], axis = 0)
DR_counts['rpavg']  = np.mean([DR_counts['rmin'], DR_counts['rmax']], axis = 0)
RR_counts['rpavg']  = np.mean([RR_counts['rmin'], RR_counts['rmax']], axis = 0)
```

```python
def plot_counts(counts, key, cut, ax, x, label):
    
    pimax = counts[key]==cut
    ax.scatter(counts[x][pimax], counts['npairs'][pimax], label = label)
```

```python
fig = plt.figure(figsize = (20, 8))
gs = gridspec.GridSpec(1, 2)

ax = fig.add_subplot(gs[0])
plot_counts(DD_counts, 'pimax', 1, ax, 'rpavg', 'DD')
plot_counts(DR_counts, 'pimax', 1, ax, 'rpavg', 'DR')
plot_counts(RR_counts, 'pimax', 1, ax, 'rpavg', 'RR')
ax.set_title('$\pi_\mathrm{{max}}={}$ $h^{{-1}}$ Mpc'.format(str(1)))
ax.set_xlabel(r'$\langle r_{p}\rangle$ $[h^{-1}$ Mpc]') # projected
ax.set_ylabel("Pair Counts")
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend()

ax = fig.add_subplot(gs[1])
plot_counts(DD_counts, 'rpavg', recenter(rbins)[-1], ax, 'pimax', 'DD')
plot_counts(DR_counts, 'rpavg', recenter(rbins)[-1], ax, 'pimax', 'DR')
plot_counts(RR_counts, 'rpavg', recenter(rbins)[-1], ax, 'pimax', 'RR')
ax.set_title(r'$\langle r_{{p}}\rangle={:.2f}$ $h^{{-1}}$ Mpc'.format(recenter(rbins)[-1]))
ax.set_xlabel('$\pi_{\mathrm{max}}$ $[h^{-1}$ Mpc]') # line of sight
ax.set_yscale('log')
plt.show()
```
